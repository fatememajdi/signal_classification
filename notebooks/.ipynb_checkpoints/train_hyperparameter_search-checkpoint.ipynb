```python
import sys
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
# Add project root to path
sys.path.append('../src')
```

```python
from train import random_search_hyperparameters
from preprocessing import load_all_signals, normalize_signal
```

```python
healthy_dir = "../data/raw/Normal Condition"
faulty_dir = "../data/raw/Looseness"
output_pkl_path = "../data/processed/signals_df.pkl"

df = load_all_signals(healthy_dir, faulty_dir, output_pkl_path)
```

```python
df['processed_signal'] = df['signal'].apply(normalize_signal)
```

```python
X = np.stack(df['processed_signal'].values)
y = df['label'].values
X, y = shuffle(X, y, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
random_search_hyperparameters(X_train, y_train, X_test, y_test)
```
