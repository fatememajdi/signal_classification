```python
import sys
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
# Add project root to path
sys.path.append('../src')
```

```python
from train import random_search_hyperparameters_2d
from preprocessing import load_dataset
```

```python
healthy_path = "../data/raw/Normal Condition"
faulty_path = "../data/raw/Looseness"
output_pkl_path = "../data/processed/signals_2d_df.pkl"

df = load_dataset(healthy_path, faulty_path, output_pkl_path)
X = np.stack(df['spectrogram'].values)[..., np.newaxis]
y = df['label'].values

X, y = shuffle(X, y, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```
